```python
import requests
import pandas as pd
from datetime import datetime, timedelta

# Define the base URL and headers
base_url = "https://www.nseindia.com/api/historical/cm/equity?symbol={symbol}&from={date}to{date}"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept': 'application/json, text/javascript, */*; q=0.01',
    'Accept-Language': 'en-US,en;q=0.9',
    'Referer': 'https://www.nseindia.com/',
    'X-Requested-With': 'XMLHttpRequest',
    'Connection': 'keep-alive'
}

# Function to generate date range
def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days) + 1):
        yield start_date + timedelta(n)

# Define the symbols and date range
symbols = ar  # Example symbols, replace with your list

# Create a session to persist certain parameters across requests
session = requests.Session()

# Send an initial request to establish the session
session.get("https://www.nseindia.com", headers=headers)

# Loop over symbols and dates to fetch data
for symbol in symbols:
    for year in range(2015, 2024):
                try:
                    start_date = datetime(year, 1, 1)
                    end_date = datetime(year, 1, 1)
                except ValueError:
                    continue  # Skip invalid dates

                for single_date in daterange(start_date, end_date):
                    date_str = single_date.strftime("%d-%m-%Y")
                    url = base_url.format(symbol=symbol, date=date_str)
                    
                    response = session.get(url, headers=headers)
                    
                    # Check if the request was successful
                    if response.status_code == 200:
                        try:
                            # Parse the JSON data
                            data = response.json()
                            
                            # If the data is structured as a list of records, convert it to a DataFrame
                            if 'data' in data:
                                df = pd.DataFrame(data['data'])
                                print(f"Data for {symbol} on {date_str}")
                                print(df.head())  # Print the first few rows of the DataFrame
                            else:
                                print(f"The JSON structure is not as expected for {symbol} on {date_str}.")
                        except ValueError:
                            print(f"Failed to parse JSON for {symbol} on {date_str}.")
                    else:
                        print(f"Failed to fetch data for {symbol} on {date_str}. Status code: {response.status_code}")
```
